```python
import json
import os
import openai

class PromptEngine:
    def __init__(self, folder_path=None, openai_api_key=None):
        # Initialize with an empty context
        self.context = []
        self.openai_api_key = openai_api_key
        openai.api_key = openai_api_key

        # Load all JSON files from the specified folder
        if folder_path:
            self.load_context_from_folder(folder_path)

    def load_context_from_folder(self, folder_path):
        """Load Q&A context from all JSON files in a folder."""
        for filename in os.listdir(folder_path):
            if filename.endswith(".json"):
                filepath = os.path.join(folder_path, filename)
                self.load_context_from_file(filepath)

    def load_context_from_file(self, filepath):
        """Load Q&A context from a single JSON file."""
        try:
            with open(filepath, "r") as file:
                qa_pairs = json.load(file)
                for qa in qa_pairs:
                    self.add_context(qa["question"], qa["answer"])
            print(f"Loaded context from file: {filepath}")
        except Exception as e:
            print(f"Error loading context from file {filepath}: {e}")

    def add_context(self, question, answer):
        """Add a new Q&A pair to the context."""
        self.context.append({"question": question, "answer": answer})

    def generate_prompt(self, user_prompt):
        """Generate a prompt with the current context and user prompt."""
        prompt = "Here is some context:\n"
        for qa in self.context:
            prompt += f"Q: {qa['question']}\nA: {qa['answer']}\n"

        # Add the user's question as the final part of the prompt
        prompt += f"\nUser's question: {user_prompt}\nAnswer:"
        return prompt

    def get_response_from_openai(self, user_prompt, model="gpt-4o-mini"):
        """Generate a response using OpenAI API with context."""
        prompt = self.generate_prompt(user_prompt)

        # Call the OpenAI API
        try:
            client = openai.OpenAI()
            messages= [{'role':'user', 'content':'tell me a joke'}]
            response = client.chat.completions.create(model='gpt-4o-mini', messages=prompt)
            
            # Retrieve and return the response text
            print(response.choices[0].message.content)
        except Exception as e:
            print(f"Error calling OpenAI API: {e}")
            return None



```
